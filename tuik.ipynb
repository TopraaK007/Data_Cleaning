```python
import pandas as pd;import re
```

```python
df=pd.read_csv(r"C:\Users\hp\OneDrive\Masaüstü\datacleaning\pivot.csv",sep="|",skiprows=4,header=None)
df
```

```python
col=["Year","Month","Product","Price","e"]
df.columns=col
```

```python
df
```

```python
df.drop(axis=1,columns=["e"],inplace=True)
```

```python
df=df.fillna(method="ffill")
```

```python
df
```

```python
df["Product"]=df["Product"].str.replace(r"\d+\.","",regex=True) #regex=True parametresini kullanarak düzenli ifadeleri etkinleştirebilirsiniz.böylece sayıları silmek için \d+ ifadesini kullanabiliyoruz
df["Product"]=df["Product"].apply(lambda x: x.replace("((","").replace(")",""))
```

```python
df["Year"]=df["Year"].astype(int)
df["Month"]=df["Month"].apply(lambda x: int(str(x)[:2]))
```
